# Variable Explaination

1. age: Age of the patient in years.
2. sex: Sex of the patient (1 = male; 0 = female).
3. cp: Chest pain type (4 values):
   0: Typical angina
   1: Atypical angina
   2: Non-anginal pain
   3: Asymptomatic
4. trestbps: Resting blood pressure (in mm Hg on admission to the hospital).
5. chol: Serum cholesterol in mg/dl.
6. fbs: Fasting blood sugar > 120 mg/dl (1 = true; 0 = false).
7. restecg: Resting electrocardiographic results (values 0, 1, 2):
   0: Normal
   1: Having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
   2: Showing probable or definite left ventricular hypertrophy by Estes' criteria
8. thalach: Maximum heart rate achieved.
9. exang: Exercise-induced angina (1 = yes; 0 = no).
10. oldpeak: ST depression induced by exercise relative to rest      (numeric value in millimeters).
11. slope: The slope of the peak exercise ST segment (values 0, 1, 2):
    0: Upsloping
    1: Flat
    2: Downsloping
12. ca: Number of major vessels (0-3) colored by fluoroscopy.
13. thal: Thalassemia (3 = normal; 6 = fixed defect; 7 = reversible
    defect).
14. target: Diagnosis of heart disease (1 = yes; 0 = no).

This dataset is commonly used to predict the presence of heart disease in a patient, with the target variable being the outcome that indicates whether the patient has heart disease (1) or not (0).

### 4. People with which chest pain are most prone to have heart disease?

```python
# 4. People with which chest pain are most prone to have heart disease?

# Creating a cross-tabulation (crosstab)
cross_tab = pd.crosstab(Main_Dataset['cp'], Main_Dataset['target'])
print("Cross-tabulation (crosstab):\n", cross_tab)

# Plotting the countplot
plt.figure(figsize = (13, 6))
sns.countplot(x ='cp', hue = 'target', data = Main_Dataset)
plt.title('Heart Disease by Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.legend(title = 'Heart Disease', labels = ['No', 'Yes'])
plt.show()
```

### 5. What is the distribution of age among people with and without heart disease?

```python
# 5. What is the distribution of age among people with and without heart disease?

crosstab = pd.crosstab(Main_Dataset['age'], Main_Dataset['target'])
print(crosstab)

plt.figure(figsize=(13, 6))
sns.histplot(data=Main_Dataset, x='age', hue='target', multiple='stack', kde=True)
plt.title('Age Distribution by Heart Disease')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
# From the above Visualization of Q.5:
# - Generally, heart disease (target 1) is more prevalent in older age groups, particularly from age 41 onwards.
# - Age groups with notably higher counts of heart disease cases include those around 54, 57, and 58 years.
# - Conversely, younger age groups, such as 29, 34, and 37 years, show very few cases of heart disease.

# Insights:
# - This distribution suggests that age could be a significant factor in determining the likelihood of heart disease.
```

### 6. What is the average cholesterol level among people with and without heart disease?

```python
# 6. What is the average cholesterol level among people with and without heart disease?

# Cross-tabulation
# Categorize cholesterol levels into bins
chol_bins = pd.cut(Main_Dataset['chol'], bins = [0, 200, 239, 500], labels = ['Desirable', 'Borderline high', 'High'])
Main_Dataset['chol_category'] = chol_bins

# Creating the cross-tabulation
cross_tab = pd.crosstab(Main_Dataset['chol_category'], Main_Dataset['target'])
print("Cross-tabulation (crosstab):\n", cross_tab)

# Violin Plot
plt.figure(figsize=(13, 6))
sns.violinplot(x = 'target', y = 'chol', data = Main_Dataset, hue = 'target', palette = ["#ff9999","#66b3ff"], dodge = False)
plt.title('Cholesterol Level by Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Cholesterol')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
# From the above visualization of Q.8, displays Higher cholesterol levels generally correlate with a higher incidence of the target condition.
```

### 7. Is there a relationship between exercise-induced angina (exang) and heart disease?

```python
# 7. Is there a relationship between exercise-induced angina (exang) and heart disease?

# Creating the cross-tabulation (crosstab) table
crosstab = pd.crosstab(Main_Dataset['exang'], Main_Dataset['target'], margins=True)
print(crosstab)

# Plotting the count plot
plt.figure(figsize=(13, 6))
sns.countplot(x='exang', hue='target', data=Main_Dataset)
plt.title('Exercise-Induced Angina by Heart Disease')
plt.xlabel('Exercise-Induced Angina')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
# From the above Visualization of Q.7:
# - Individuals without exercise-induced angina (exang=0) constitute the majority in both heart disease groups: 225 without heart disease and 455 with heart disease.
# - Conversely, individuals with exercise-induced angina (exang=1) are predominantly represented in the group without heart disease (274), while only 71 have heart disease.
# - Overall, the majority of individuals (680) do not experience exercise-induced angina, compared to 345 who do.
```

### 8. How does the slope of the peak exercise ST segment relate to the presence of heart disease?

```python
# 8. How does the slope of the peak exercise ST segment relate to the presence of heart disease?

# Creating the cross-tabulation (crosstab) table
crosstab = pd.crosstab(Main_Dataset['slope'], Main_Dataset['target'], margins=True)
print(crosstab)

# Plotting the count plot
plt.figure(figsize=(13, 6))
sns.countplot(x='slope', hue='target', data=Main_Dataset)
plt.title('Slope of Peak Exercise ST Segment by Heart Disease')
plt.xlabel('Slope of Peak Exercise ST Segment')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
# From the above Visualization of Q.8:
# - Individuals with a slope of 1 (upsloping) are the most prevalent in both heart disease groups: 324 without heart disease and 158 with heart disease.
# - Slope 2 (flat) shows a higher prevalence of heart disease cases (340) compared to those without heart disease (129).
# - Slope 0 (downsloping) has the lowest counts in both heart disease groups: 46 without heart disease and 28 with heart disease.
```
